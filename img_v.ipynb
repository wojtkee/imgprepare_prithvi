```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
parcel_id=np.load(r'F:\mgr_datasets\PASTIS\ANNOTATIONS\ParcelIDs_10001.npy')
heatmap=np.load(r'F:\mgr_datasets\PASTIS\INSTANCE_ANNOTATIONS\HEATMAP_10001.npy')
instances=np.load(r'F:\mgr_datasets\PASTIS\INSTANCE_ANNOTATIONS\INSTANCES_10001.npy')
zones=np.load(r'F:\mgr_datasets\PASTIS\INSTANCE_ANNOTATIONS\ZONES_10001.npy')
target=np.load(r'F:\mgr_datasets\PASTIS\ANNOTATIONS\TARGET_10001.npy')
data=np.load(r'F:\mgr_datasets\PASTIS\DATA_S2\S2_10001.npy')
```

```python
def normalize(img):
    img = ((img) / np.max(img)) * 255
    img = img.astype(np.uint8)
    return img
```

```python
plt.imshow(normalize(parcel_id))
plt.show()
plt.imshow(normalize(heatmap))
plt.show()
plt.imshow(normalize(instances))
plt.show()
plt.imshow(normalize(zones))
plt.show()
plt.imshow(normalize(target[0,:,:]))
plt.show()
plt.imshow(normalize(target[1,:,:]))
plt.show()
plt.imshow(normalize(target[2,:,:]))
plt.show()

plt.imshow(normalize(data[0,0,:,:]))
plt.show()
```
